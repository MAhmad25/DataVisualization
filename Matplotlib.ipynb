```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,3)
y=x**2
print(x,y)
```

```python
plt.plot(x,y)
plt.title("Linear Distribution")
plt.xlabel("X Values")
plt.ylabel("Y Values")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

<h3>Object Oriented way</h3>

```python
fig=plt.figure()
axis1=fig.add_axes([0,0,1,1])
axis1.plot(x,y)
axis1=fig.add_axes([0.3,0.6,.3,.3])
axis1.plot(y,x)
```
