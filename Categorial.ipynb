```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=sb.load_dataset("tips")
df
```

```python
#Find how many men and women are smoker or not
#Count Plot
sb.countplot(x=df["sex"],hue=df["smoker"])
```

```python
#Bar Plot
import numpy as np
sb.barplot(x=df["sex"],y=df["total_bill"],estimator=np.sum)# by defualt its uses estimator function mean aggregate
```

```python
#Box Plot
sb.boxplot(data=df,x="total_bill",y="day")
```

```python
#Violin Plot
sb.violinplot(data=df,x="day",y="total_bill")
```

```python
#Strip Plot
sb.stripplot(data=df,x="day",y="total_bill")
```

```python
#Swarm Plot
sb.swarmplot(data=df,x="day",y="total_bill")
```
