<h3>Type of Distribution Plot</h3>

```python
#using seaborn
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=sb.load_dataset("tips")
df
```

```python
df["size"].unique()# checking if size is catgorical? yes it is
```

<h3>Hist Plot</h3>

```python
plt.subplot(2,2,1)
sb.histplot(df["tip"],bins=20,kde=True) #bin=20 is default
plt.subplot(2,2,2)
sb.histplot(df["total_bill"],bins=20,kde=True)
plt.subplot(2,2,3)
sb.histplot(df["smoker"],bins=20,kde=True)
```

<h3>Joint Plot</h3

```python
sb.jointplot(x=df["total_bill"],y=df["tip"],kind="reg")
```

<h3>Pair Plot</h3>

```python
sb.pairplot(df,hue="sex")
```

```python
sb.rugplot(df)
```
